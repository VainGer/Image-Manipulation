### Import the required libraries - 2 points

````python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2
#google colab
from google.colab import files
````

### Plot the raw image - 3 points

````python
#google colab
raw_img = files.upload()

raw_image = Image.open("dog.jpg")
plt.imshow(raw_image)
plt.axis('off')
plt.show()
````

### Apply face detection and print the outcome - 10 points

````python
face_cascade = cv2.CascadeClassifier("dog_face.xml")
cv_img = cv2.cvtColor(np.array(raw_image), cv2.COLOR_RGB2BGR)
gray = cv2.cvtColor(cv_img, cv2.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=8)
print(f"Number of faces detected: {len(faces)}")
print(f"Face position: x={faces[0][0]}, y={faces[0][1]}")
print(f"Face size: w={faces[0][2]}, h={faces[0][3]}")
````

### Plot the image with the detected face - 5 points

````python
x,y,w,h = faces[0]
max_area = w*h
for(x2,y2,w2,h2) in faces:
  area = w2*h2
  if area > max_area:
    x,y,w,h = x2,y2,w2,h2

image_with_box = cv_img.copy()
cv2.rectangle(image_with_box, (x, y), (x + w, y + h), (0, 0, 255), 10)

plt.imshow(cv2.cvtColor(image_with_box, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Detected Dog Face")
plt.show()
````

### Image manipulation 1: blur the entire background and leave face unchanged, or vice versa - 10 points

**Important Note:** you should apply the filters manually on the pixels and not use pre-defind functions. doing so will result deduction in points.<br>
You should explain the function you wrote in your own words, that is 4 points.

````python
image_np = np.array(raw_image)

kernel_size = 19
pad = kernel_size // 2

height, width, channel = image_np.shape
for i in range (height):
  for j in range (width):
    if i < y or i > y + h or j < x or j > x + w:
      for c in range(channel):
          y_start = max(0, i - pad)
          y_end = min(height, i + pad + 1)
          x_start = max(0, j - pad)
          x_end = min(width, j + pad + 1)
          kernel_area = image_np[y_start:y_end, x_start:x_end, c]
          image_np[i, j, c] = kernel_area.mean()

image_blurred = Image.fromarray(image_np)
plt.imshow(image_blurred)
plt.axis('off')
plt.title("Blurred background")
plt.show()
````

What you did in the function? explain:

1. created the array of the image.
2. Configured the kernel size and calculated the padding for colors average calculations.
3. Iterated through each pixel of the image, except the detected face area.
4. In channel loop for each pixel i found its kernel area except the array boundries(from 0 to array length) the y/x starts and ends so the kernel area is a rectangular area from the original array.
5. Assigned the calculated average from the kernel(the .mean() function) to the image array.
6. Converted the array back to image.
7. Plot the image

>**The kernel size was exagerated on purpose, because the original image was blurred a bit already, so it made it more visible.**

### Image manipulation 2: pixelate the entire background and leave face unchanged, or vice versa - 10 points

**Important Note:** you should apply the filters manually on the pixels and not use pre-defind functions. doing so will result deduction in points.<br>
You should explain the function you wrote in your own words, that is 4 points.

````python
image_np = np.array(raw_image)

height, width, channel = image_np.shape

block_size = 20

for i in range(0, height, block_size):
    for j in range(0, width, block_size):
         if i < y or i > y + h or j < x or j > x + w:
               block = image_np[i:i + block_size, j:j + block_size]
               avg_pixel = block.mean(axis=(0, 1))
               image_np[i:i + block_size, j:j + block_size] = avg_pixel

image_pixelated = Image.fromarray(image_np)
plt.imshow(image_pixelated)
plt.axis('off')
plt.title("Pixelated background")
plt.show()
````

What you did in the function? explain:

1. Created the array of the image.
2. Configured the block area size for pixelation.
3. Iterated through each block area except the detected face area.
4. Created an array of each block based on the block size.
5. Calculated average pixel color for each block, the axis = (0,1) calculates the average for a 2D area of array so the color information retained.
6. Converted the array back to image.
7. Plot the image

### Image manipulation 3: make the entire background **negative** and leave face unchanged, or vice versa - 10 points

**Important Note:** you should apply the filters manually on the pixels and not use pre-defind functions. doing so will result deduction in points.<br>
You should explain the function you wrote in your own words, that is 4 points.

````python
image_np = np.array(raw_image)

height, width, channel = image_np.shape
for i in range (height):
  for j in range (width):
    if i < y or i > y + h or j < x or j > x + w:
      for c in range(channel):
          image_np[i,j,c] = 255 - image_np[i,j,c]

image_negative = Image.fromarray(image_np)
plt.imshow(image_negative)
plt.axis('off')
plt.title("Negative background")
plt.show()
````

### Image manipulation 4: search and pick a new filter and apply it to the background. leave face unchanged, or vice versa - 10 points

**Important Note:** you should apply the filters manually on the pixels and not use pre-defind functions. doing so will result deduction in points.<br>
You should explain the function you wrote in your own words, that is 4 points.

````python
image_np = np.array(raw_image)
height, width, channel = image_np.shape
for i in range (height):
  for j in range (width):
    if i < y or i > y + h or j < x or j > x + w:
      r, g, b = image_np[i, j]
      new_r = 0.393 * r + 0.769 * g + 0.189 * b
      new_g = 0.349 * r + 0.686 * g + 0.168 * b
      new_b = 0.272 * r + 0.534 * g + 0.131 * b
      image_np[i, j] = [min(255, new_r), min(255, new_g), min(255, new_b)]

image_sepia = Image.fromarray(image_np)
plt.imshow(image_sepia)
plt.axis('off')
plt.title("Sepia background")
plt.show()
````

What you did in the function? explain:

I decided to apply sepia tone filter to the background, i iterated through the image array and calculated new values for each pixel based on formula i found on https://stackoverflow.com/questions/1061093/how-is-a-sepia-tone-created
This formula takes original values of RGB and calculates new values as:

New Red = (Original Red √ó 0.393) + (Original Green √ó 0.769) + (Original Blue √ó 0.189)

New Green = (Original Red √ó 0.349) + (Original Green √ó 0.686) + (Original Blue √ó 0.168)

New Blue = (Original Red √ó 0.272) + (Original Green √ó 0.534) + (Original Blue √ó 0.131)

if one of this new values exceeding 255, then the function will apply 255 as a max value for a color
